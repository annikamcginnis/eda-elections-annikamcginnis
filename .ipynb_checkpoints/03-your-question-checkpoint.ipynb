### My original question was this "Did counties where Democrats flipped senate voting majorities between 2020 and 2024 vote for Trump or Harris in the presidential race?" - but during my analysis, I realized we only have Senate data for 2024.

## I switched to this question: "We all know Democrats didnâ€™t do too well in the 2024 election. But where did Democrats actually do better in presidential elections in 2024 compared to 2020? In these counties, was the Democratic presidential or senate candidate more popular? 

Write code below to modify/pivot/etc the dataframe to answer your question. Perform any calculations you need.

```python
df = election_results[(election_results['year'] == 2024) | (election_results['year'] == 2020)]
```

```python
df = df[['fips', 'type', 'state', 'county', 'year', 'total', 'democratic_pct', 'republican_pct', 'other']]
df
```

```python
df = df.pivot_table(index=['fips', 'state', 'county'], columns=['type', 'year'])
df
```

```python
df.columns = df.columns.map(lambda x: '_'.join(map(str, x)) if isinstance(x, tuple) else str(x))
df = df.reset_index()
```

```python
df = df.drop('index', axis=1)
```

```python
df
```

```python
df['dem_pres_pct_diff'] = df['democratic_pct_presidential_2024'] - df['democratic_pct_presidential_2020']
df.head()
```

```python
dem_better = df[df['dem_pres_pct_diff'] > 0]
dem_better
```

```python
df['rep_pres_pct_diff'] = df['republican_pct_presidential_2024'] - df['republican_pct_presidential_2020']
df.head()
```

```python
rep_better = df[df['rep_pres_pct_diff'] > 0]
rep_better
```
